#### Zadanie 1 
Zaimplemenuj metodę eliminacji Gaussa bez pivotingu i z pivotingiem dla układu równań o dowolnym rozmiarze.

```python
def gauss_elimination(A, B):
    n = len(A)

    if len(A) != len(A[0]):
        return "invalid matrix"

    if A[0][0] == 0:
        return "division by zero will occur"

    for row in range(n - 1):
        for i in range(row + 1, n):
            factor = A[i][row] / A[row][row]
            for j in range(row, n):
                A[i][j] -= factor * A[row][j]
            B[i] -= factor * B[row]

    result = [0] * n
    result[n - 1] = B[n - 1] / A[n - 1][n - 1]
    for row in range(n - 2, -1, -1):
        sum = B[row]
        for j in range(row + 1, n):
            sum -= A[row][j] * result[j]
        result[row] = sum / A[row][row]
    return result


def gauss_elimination_pivoting(A, B):
    n = len(A)

    if len(A) != len(A[0]):
        return "invalid matrix"

    M = A
    i = 0

    for x in M:
        x.append(B[i])
        i += 1

    for row in range(n):
        for i in range(row, n):
            if abs(M[i][row]) > abs(M[row][row]):
                M[row], M[i] = M[i], M[row]
            else:
                pass

        for i in range(row + 1, n):
            factor = M[i][row] / M[row][row]
            for j in range(row, n + 1):
                M[i][j] -= factor * M[row][j]

    x = [0] * n

    x[n - 1] = M[n - 1][n] / M[n - 1][n - 1]
    for i in range(n-1, -1, -1):
        z = 0
        for j in range(i + 1, n):
            z += float(M[i][j]) * x[j]
        x[i] = float(M[i][n] - z)/M[i][i]
    return x
```

#### Zadanie 2
Zademonstruj działanie algorytmu na macierzy o rozmiarze 5 x 5. Zademonstruj w jakiej sytuacji potrzebny jest pivoting i jak działa. 

```python
from tabulate import tabulate

A = [[10, -7, 0, 12, 24],
     [-3, 2, 6, 54, 2],
     [5, -1, 5, 432, 21, 1],
     [23, 532, 12, 32, 4],
     [53, 23, 643, 1, 432]]
B = [7, 3, 6, 342 ,1]
x1 = gauss_elimination(A, B)

A = [[10, -7, 0, 12, 24],
     [-3, 2, 6, 54, 2],
     [5, -1, 5, 432, 21, 1],
     [23, 532, 12, 32, 4],
     [53, 23, 643, 1, 432]]
B = [7, 3, 6, 342 ,1]
x2 = gauss_elimination_pivoting(A, B)

A = [[0.00000000000000012, 4200, 34],
     [0.5, 69, 2320],
     [1, 17, 7]]
B = [7, 3.901, 6]
x3 = gauss_elimination(A, B)

A = [[0.00000000000000012, 4200, 34],
     [0.5, 69, 2320],
     [1, 17, 7]]
B = [7, 3.901, 6]
x4 = gauss_elimination_pivoting(A, B)

A = [[0, 4200, 34],
     [0.5, 69, 2320],
     [1, 17, 7]]
B = [7, 3.901, 6]
x5 = gauss_elimination(A, B)

A = [[0, 4200, 34],
     [0.5, 69, 2320],
     [1, 17, 7]]
B = [7, 3.901, 6]
x6 = gauss_elimination_pivoting(A, B)

table = [[x1, x2],
        [x3, x4],
        [x5, x6]]
print(tabulate(table, headers=["gauss elimination", "gauss elimination with pivoting"], tablefmt="fancy_grid"))
```

Pivoting polega na wyszukaniu w danej kolumnie macierzy podczas iteracji eliminacji Gaussa największej co do wartości bezwzględnej liczby (pivota) oraz zamianie wierszy - tego w którym znaleziony pivot się znajduje oraz wiersza który zawiera liczbę znajdującą się na przecięciu diagonali oraz obecnie rozważanej kolumny. Po tej operacji na diagonali umieszczony zostaje pivot - skutkuje to zwiększeniem dokładności obliczeń, gdyż dzięki temu zabiegowi w następnych etapach algorytmu nie nastąpi mnożenie przez stosunkowo duże liczby, które spowodowałoby zwiększenie błędu.

Pivoting potrzebny jest w sytuacji gdy:
- w danej iteracji eliminacji gaussa na przęcięciu diagonali i rozważanej kolumny znajduje się 0 - eliminacja gaussa bez pivotingu przestaje działać w momencie, gdy w tej samej kolumnie pod diagonalą znajdują sie liczby różne od 0 (następuje dzielenie przez 0)
- w danej iteracji eliminacji gaussa na przecięciu diagonali i rozważanej kolumny znajduje się bardzo mała liczba - eliminacja bez pivotingu uzna ją za element podstawowy i przeprowadzi eliminację przy jego użyciu w rezultacie propagując błąd spowodowany mnożeniem przez dużo większe liczby w następnych krokach algorytmu. 

#### Zadanie 3
Podaj teorytyczną złożoność obliczeniową algorytmu eliminacji Gaussa. Przeprowadź testy wydajności swojego algorytmu sprawdzając jego działanie dla różnych rozmiarów macierzy (testy powinny być wykonane poza środowiskiem jupyter). Aby wygenerować układ równań, wygeneruj wektor rozwiązań i macierz współczynników losując wartości (skorzystaj z funkcji poznanych w Ćwiczeniu 2) i następnie oblicz wektor wyrazów wolnych. 

Teoretyczna złożoność obliczeniowa dla eliminacji Gaussa: O(n^3)
Do przeprowadzenia testów wykonałem poza środowiskiem jupyter użyłem poniższego kodu:

```python
import copy
import time
from random import randrange
import numpy as np
import matplotlib.pyplot as plt


def randomVector(n):
    return np.random.rand(n)

def randomMatrix(n):
    return np.random.rand(n, n)

def measure_time():
    n = randrange(100, 420)
    A = randomMatrix(n).tolist()
    B = randomVector(n).tolist()

    start = time.time()
    gauss_elimination(copy.deepcopy(A), copy.deepcopy(B))
    stop = time.time()
    gauss = stop - start

    start = time.time()
    gauss_elimination_pivoting(copy.deepcopy(A), copy.deepcopy(B))
    stop = time.time()
    gauss_pivoting = stop - start

    return gauss, gauss_pivoting, n

if __name__ == '__main__':
    gauss = []
    gauss_pivoting = []
    size = []
    for i in range(100):
        gauss_time, gauss_pivoting_time, n = measure_time()
        gauss.append(gauss_time)
        gauss_pivoting.append(gauss_pivoting_time)
        size.append(n)

    for i in range(100):
        f = open("file.txt", "a")
        f.write(f"matrix size: {size[i]} \t gauss elimination: {gauss[i]}[s] \t gauss elimination with pivoting: {gauss_pivoting[i]}[s]"'\n')
        f.close()

    plt.plot(size, gauss, 'g.', label="gauss")
    plt.plot(size, gauss_pivoting, 'r.', label="gauss pivoting")
    plt.xlabel("matrix size")
    plt.ylabel("time[s]")
    plt.title("gauss elimination time measurement")
    plt.legend()
    plt.grid(True)
    plt.savefig("gauss_plot")
    plt.show()
```
